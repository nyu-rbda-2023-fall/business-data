```python
### Upload heatmapcsv csv then run through sections to get world heatmap
#Reference: plotly.com Density heatmap guide used for heatmap 
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
import plotly.graph_objects as go
df2 = pd.read_csv(io.BytesIO(uploaded['heatmapcsv']))
```

```python
fig = go.Figure(go.Densitymapbox(lat=df2.Latitude, lon=df2.Longitude, z=df2.review_count,
                                 radius=10))
fig.update_layout(mapbox_style="stamen-terrain", mapbox_center_lon=180)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
